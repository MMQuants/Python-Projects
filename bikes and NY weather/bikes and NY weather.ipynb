In order to save time and reuse code I will create a method that will clean and prepare the data for me and then I will apply it to the datasets that I have just loaded.


```python
def data_clean_and_prep(df):
    """
    Docstring
    Prepares and cleans the data by removing any char:'T' and replacing it with zero:'0'
    
    :param df: a wheather dataset that has to be prepared for further usage
    
    :return: this method returns the new cleaned and prepared dataset
    """
    # lets turn the Date column into datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    for item in df.columns[1:]:
        # column clean and prep
        if df[item].astype(str).str.contains('T').any():
            df[item] = df[item].astype(str).str.replace('T','0')
            df[item] = df[item].astype(float) 
        else:
            df[item] = df[item].astype(float)
    return df
```

```python
Bridgeport_Area_Preped = data_clean_and_prep(Bridgeport_Area_data)
Islip_Area_data_Preped = data_clean_and_prep(Islip_Area_data)
Newark_Area_data_Preped = data_clean_and_prep(Newark_Area_data)
Central_Park_Area_data_Preped = data_clean_and_prep(Central_Park_Area_data)
Kennedy_AP_Area_data_Preped = data_clean_and_prep(Kennedy_AP_Area_data)
LaGuardia_AP_Area_data_Preped = data_clean_and_prep(LaGuardia_AP_Area_data)
```

Now the New York weather datasets are clean and prepared for further usage.
Тhe next step is to merge the weather datasets into one.

```python
temp1 = pd.merge(Bridgeport_Area_Preped, Islip_Area_data_Preped, on="Date", suffixes=["_Bridgeport", "_Islip"])
temp2 = pd.merge(Newark_Area_data_Preped, Central_Park_Area_data_Preped, on="Date", suffixes=["_Newark", "_Central"])
temp3 = pd.merge(Kennedy_AP_Area_data_Preped, LaGuardia_AP_Area_data_Preped, on="Date", suffixes=["_Kennedy", "_LaGuardia"])
```

```python
hold1 = pd.merge(temp1, temp2, on="Date")
all_NY_weather_zones = pd.merge(hold1, temp3, on="Date")
```
